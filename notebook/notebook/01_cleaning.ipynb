```python
import pandas as pd
import numpy as np

    # Load the dataset
df = pd.read_csv('../raw/Pokemon.csv')
df.head()
```

```python
# Contar la cantidad de valores nulos en cada columna
null_counts = df.isnull().sum()
print(null_counts)
```

```python
# Rellenar los valores nulos con "None"
df_filled = df.fillna("None")
print(df_filled.isnull().sum())
```

```python
# Reordenar las columnas asi: pokedex_number,name,type1,type2,total,hp,attack,defense,sp_atk,sp_def,speed,generation,legendary
df_reordered = df_filled[['#', 'Name', 'Type 1', 'Type 2', 'Total', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Generation', 'Legendary']]
# Renombrar las columnas
df_renamed = df_reordered.rename(columns={
    '#': 'pokedex_number',
    'Name': 'name',
    'Type 1': 'type1',
    'Type 2': 'type2',
    'Total': 'total',
    'HP': 'hp',
    'Attack': 'attack',
    'Defense': 'defense',
    'Sp. Atk': 'sp_atk',
    'Sp. Def': 'sp_def',
    'Speed': 'speed',
    'Generation': 'generation',
    'Legendary': 'legendary'
})
df_renamed.head()
```

```python
# hacer que la columna pokedex_number no repita el mismo numero, es decir, que cada pokemon tenga un numero unico
df_renamed = df_renamed.drop_duplicates(subset=['pokedex_number'])
df_renamed.head()
```

```python
# Guardar el DataFrame limpio en un nuevo archivo CSV
df_renamed.to_csv('../processed/cleaned_pokemon.csv', index=False)
```
